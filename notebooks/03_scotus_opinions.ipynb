# SCOTUS Opinions Text Analysis

**Dataset**: 15 landmark Supreme Court majority opinions (public domain)
**Date Range**: 1801–2015
**Source**: Supreme Court of the United States
**License**: U.S. Government Works — Public Domain

**Method**: Live API / real data patterns → text length trends → topic distribution → vote margin analysis

```python
import json
from collections import Counter
import numpy as np

# Load real SCOTUS data
with open('../projects/scotus-opinions/data/scotus_cases.json') as f:
    cases = json.load(f)

print(f'Loaded {len(cases)} landmark SCOTUS opinions')
```

```python
# EDA summary
terms = [c['term'] for c in cases]
words = [c['word_count'] for c in cases]
topics = [c['topic'] for c in cases]
dispositions = [c['disposition'] for c in cases]
margins = [c['votes_for'] - c['votes_against'] for c in cases]

print(f'Term range: {min(terms)} - {max(terms)}')
print(f'Word count range: {min(words)} - {max(words)}')
print(f'Topics: {dict(Counter(topics))}')
print(f'Dispositions: {dict(Counter(dispositions))}')
print(f'Vote margins: min={min(margins)}, max={max(margins)}, mean={np.mean(margins):.1f}')
```
