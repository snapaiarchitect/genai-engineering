# PubMed Biomedical Research Analysis

**Dataset**: 20 drug trials, 12 biomarkers, 10 epidemiology records
**Source**: Real clinical trial patterns (KEYNOTE, CheckMate, IMvigor, etc.)

**Method**: Live API / real data patterns

## Figure 1: Drug Response Rates by Cancer Type

```python
import json; import matplotlib.pyplot as plt; import numpy as npwith open('data/drug_trials.json') as f: trials = json.load(f)cond_data = {}; [cond_data.setdefault(t['condition'], []).append(t['response_rate']) for t in trials]means = {k: np.mean(v) for k,v in cond_data.items()}fig, ax = plt.subplots(figsize=(12,7)); ax.bar(means.keys(), means.values(), color='teal')ax.set_title('Immunotherapy Response Rates'); plt.xticks(rotation=45); plt.show()
```

## Figure 2: Biomarker Expression Volcano Plot

```python
import json; import matplotlib.pyplot as plt; import numpy as npwith open('data/biomarkers.json') as f: bms = json.load(f)fig, ax = plt.subplots(figsize=(12,7)); ax.scatter([b['expression_fold_change'] for b in bms], [-np.log10(b['p_value']) for b in bms], s=100)ax.set_title('Biomarker Volcano Plot'); plt.show()
```

## Figure 3: Disease Burden: Prevalence vs Mortality

```python
import json; import matplotlib.pyplot as pltwith open('data/epidemiology.json') as f: epi = json.load(f)fig, ax = plt.subplots(figsize=(11,7)); ax.scatter([e['prevalence'] for e in epi], [e['mortality_rate'] for e in epi], s=200, c='green')ax.set_title('Disease Burden'); plt.show()
```
